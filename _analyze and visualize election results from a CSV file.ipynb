 ##### this code is designed to analyze and visualize election results from a CSV file,
 ##### focusing on the top parties and displaying their percentage of votes in a bar chart.
##### It allows you to customize the number of parties to display in the analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
def analyze_election_results(file_name, num_parties=None):

    df = pd.read_csv(file_name, delimiter=';', encoding='utf-8')
    
    # Group  by party code  
    party_votes = df.groupby('Partikode')['Antall stemmer totalt']
    #calculate the total votes
    party_votes= party_votes.sum().reset_index()
    
    total_votes = party_votes['Antall stemmer totalt'].sum() 
    
    # Sort parties by votes 
    sorted_parties = party_votes.sort_values(by='Antall stemmer totalt', ascending=False)
    
    # add  percentage columme  of votes received by each party 
    sorted_parties['percentage']=(sorted_parties['Antall stemmer totalt']/ total_votes)*100
    print(f"Party{'':<10} Votes{'':<10} Percentage{'':<10}")
    print("-" * 50) 
    for index,  row in sorted_parties[:num_parties].iterrows():
        if row[2]>4: # row[2] : percentage of votes received by each party 
            
            print(f'{row[0]:<15}{row[1]:<15}{row[2]:.2f} ****')
        else: print(f'{row[0]:<15}{row[1]:<15}{row[2]:.2f}')    


   
    colors = ['blue', 'green', 'red', 'purple', 'orange', 'pink', 'brown', 'gray', 'cyan', 'magenta']
    plt.figure(figsize=(12, 6))
    plt.bar(sorted_parties[:num_parties]['Partikode'], sorted_parties[:num_parties]['percentage'],color=colors)
    plt.xlabel('Party Code')
    plt.ylabel('Percentage of Votes')
    plt.title('Percentage of Votes Received by Each Party')
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
    plt.tight_layout()
    plt.show()

        

# Example usage with 3 parties
print("Results for 3 parties:")
analyze_election_results("2021-09-14_party distribution_1_st_2021.csv",3)

# Example usage with 7 parties
print("Results for 7 parties:")
analyze_election_results("2021-09-14_party distribution_1_st_2021.csv", num_parties=7)


# Example usage with all parties
print("\nResults for all parties:")
analyze_election_results("2021-09-14_party distribution_1_st_2021.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def analyze_election_results(file_name, num_parties=None):

    df = pd.read_csv(file_name, delimiter=';', encoding='utf-8')
    
    # Group by party code  
    party_votes = df.groupby('Partikode')['Antall stemmer totalt']
    # Calculate the total votes
    party_votes = party_votes.sum().reset_index()
    
    total_votes = party_votes['Antall stemmer totalt'].sum() 
    
    # Sort parties by votes 
    sorted_parties = party_votes.sort_values(by='Antall stemmer totalt', ascending=False)
    
    # Add a percentage column of votes received by each party 
    sorted_parties['percentage'] = (sorted_parties['Antall stemmer totalt'] / total_votes) * 100

    # Define colors for the bars (you can customize these colors)
    colors = ['blue', 'green', 'red', 'purple', 'orange', 'pink', 'brown', 'gray', 'cyan', 'magenta']

    plt.figure(figsize=(12, 6))
    bars = plt.bar(sorted_parties[:num_parties]['Partikode'], sorted_parties[:num_parties]['percentage'], color=colors)
    plt.xlabel('Party Code')
    plt.ylabel('Percentage of Votes')
    plt.title('Percentage of Votes Received by Each Party')
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability

    # Add labels above each bar with the party's name
    for bar, party_name in zip(bars, sorted_parties[:num_parties]['Partikode']):
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 1, party_name, ha='center')

    plt.tight_layout()
    plt.show()

# Example usage with 3 parties
print("Results for 3 parties:")
analyze_election_results("2021-09-14_party distribution_1_st_2021.csv", 3)

# Example usage with 7 parties
print("Results for 7 parties:")
analyze_election_results("2021-09-14_party distribution_1_st_2021.csv", num_parties=7)

# Example usage with all parties
print("\nResults for all parties:")
analyze_election_results("2021-09-14_party distribution_1_st_2021.csv")
```
